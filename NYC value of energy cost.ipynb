**Question 2:** 
What is the total number of jobs created for businesses in Queens?

```python
total_num_job_created_queens = df[df['City']=='Queens']['Job created'].fillna(0).sum()
print(total_num_job_created_queens)
```

**Question 3:**
How many different unique email domains names are there in the data set?

```python
df['company email'].fillna('',inplace=True)
```

```python
split_email = [x.split('@') if x!='' else '' for x in df['company email']]
```

```python
len(split_email)
```

```python
unformated_email = [x for x in split_email if len(x)!=2]
```

```python
unformated_email
```

```python
for x in range(len(unformated_email)):
    split_email.remove(unformated_email[x])
```

```python
domain_email = [x[1] for x in split_email]
domain_email
```

```python
num_unique_domains_names = len(pd.Series(domain_email).unique())
print(num_unique_domains_names)
```

**Question 4:**
Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
filtered = df.groupby(['NTA']).filter(lambda x:x['Company Name'].count()>=5)
r1 = filtered.groupby(['NTA']).sum()['Job created'].reset_index()
r2 = filtered.groupby(['NTA']).mean()['Total Savings'].reset_index()
result = r1.merge(r2, left_on='NTA', right_on='NTA')
result.filter(['Job created','Total Savings'])
result.rename(columns={"Job created": "total_jobs_created", "Total Savings": "average_total_savings"}, inplace=True)
```
